```python
import numpy as np
import matplotlib.pyplot as plt
import math
from numpy import random

mean,variance = 5,1
sigma = math.sqrt(variance)
```

```python
def gaussianDistribution(mean,variance,a):
    return 1/(math.sqrt(2*math.pi)*np.sqrt(variance)) * math.exp(-(a - mean)**2/(2*np.sqrt(variance)**2))
```

```python
def likelihood(a, mean):
    l = 1
    for i in a:
        l *= gaussianDistribution(mean,1,i)
    return l
```

```python
a = random.normal(mean,variance, 10)
y = [gaussianDistribution(mean,variance,i) for i in a]
plt.plot(a, y, 'ro')
plt.title("Gaussian Distribution")
```

```python
Means = [0]*11
Likelihoods = [0]*11
for i in range(0, 11):
    Likelihoods[i] = likelihood(a, i)
    Means[i] = i
plt.plot(Means, Likelihoods, 'r')
plt.xlabel("Mean")
plt.ylabel("Likelihood")
plt.title("Likelihood curve")
```

```python
log_likelihood = [math.log(i) for i in Likelihoods]
plt.plot(means,log_likelihood,'r')
plt.xlabel("Mean")
plt.ylabel("Log_Likelihood")
plt.title("Log Likelihood curve")
```
